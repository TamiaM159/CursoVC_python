Reproducir video desde archivo

```python
import numpy as np
import cv2 as cv

# crear un objeto VideoCapture su parametro sera la direccion del video
cap = cv.VideoCapture("video.mp4")
# creamos un bucle

while(cap.isOpened()):
    
    ret,frame = cap.read()
    if ret:
        
        cv.imshow("TM", frame)
        if(cv.waitKey(10) & 0xFF ==ord("q")):
            break
    else:
        break
cap.release()
cv.destroyAllWindows()
```

Funci√≥n para Dibujar Geometrias en OpenCV
Dibujar linea

```python
# la funcion cv.line()

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# crear una imagen
img = np.zeros((600,700,3), np.uint8)
# dibujamos la linea en la imagen
cv.line(img,(100,150),(450,450),(255,0,0),6)
# mostrar la imagen
plt.imshow(img)
```

Dibujar circulo

```python
# la funcion cv.circle()
# Parametros: img, condenadas centrales, radio, color, grosor)
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# crear una imagen
img = np.zeros((520,640,3), np.uint8)
# dibujamos circulo
cv.circle(img,(300,250), 100,(255,0,100),-1)
#mostrar nuestra imagen
plt.imshow(img)
```

Captura de eventos desde el mouse

```python
import cv2 as cv

clicked = False

# funcion encargada de capturar el click del mouse
def onMouse(evento,x,y,flags,param):
    # variable global
    global clicked
    if(evento ==cv.EVENT_LBUTTONUP):
        clicked =True
# creamos objeto cv.VideoCapture()
camCap = cv.VideoCapture(0)
cv.namedWindow("MyWindow")
# configuramos se envie a la funcion onMouse los eventos del mouse
cv.setMouseCallback("MyWindow",onMouse)
print("Mostrando estado de la Camara, Presione en la Ventana para detener")
ret,frame = camCap.read()
while(ret and cv.waitKey(1)==-1 and not clicked):
    cv.imshow("MyWindow", frame)
    ret,frame = camCap.read()
camCap.release()
cv.destroyAllWindows()
```

Ejemplo: Uso de eventos con el rat√≥n

```python
import numpy as np
import cv2 as cv
def draw_circle(event, x, y, flags, param):
    if(event == cv.EVENT_LBUTTONDBLCLK):
        cv.circle(img, (x, y), 25, (255,0,0), -1 )
img = np.zeros((512,512,3), np.uint8)
cv.namedWindow('Dibujando Circulos')
cv.setMouseCallback('Dibujando Circulos', draw_circle)
while(True):
    cv.imshow('Dibujando Circulos',img)
    if (cv.waitKey(1) & 0xFF ==ord('t')):
        break
cv.destroyAllWindows()
```
